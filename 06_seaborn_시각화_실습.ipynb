## scatterplot
* hue : 색깔로 구분
* size : 데이터의 크기 표현

```python
# -----------------------------------
# 지불금액(total_bill)에 따른 팁(tip)
#   - 점의 색으로 성별 구분
#   - 점의 크기로 테이블 인원수(size) 구분
# -----------------------------------

# 👉 여기에 코드를 작성하세요
plt.figure(figsize=(7,7))
sns.scatterplot(tips, x='total_bill', y = 'tip', hue='sex', size = 'size') #컬럼 이름으로 라벨 이름까지 자동으로 만들어준다.(seaborn최고!) 하지만 라벨바꾸고 싶다면
plt.xlabel('지불금액')
plt.ylabel('팁')
```

## countplot
* 데이터의 빈도수 시각화
* `sns.countplot(data=데이터프레임, x='범주형컬럼')`

```python
# ---------------------
# titanic - who의 빈도수
# ---------------------

# 👉 여기에 코드를 작성하세요
plt.figure(figsize=(15,5))
sns.countplot(data=titanic, x = 'age')
plt.xticks(rotation=90);              # countplot은 histplot과 달리 구간이 아니라, 각 x축 data의 값을 전부 나타냄. 그러니까 '범주data를 넣는게 좋음(범주는 몇개 안되니까)'
```

```python
# ---------------------
# titanic - who의 빈도수 시각화
# ---------------------

# 👉 여기에 코드를 작성하세요
sns.countplot(data=titanic, x = 'who')   # 이처럼 범주를 넣을 때 사용한다.
```

```python
# ---------------------
# titanic - who의 빈도수 시각화 결과에서
# 각 범주를 한번 더 나누기 ()
# ---------------------

# 👉 여기에 코드를 작성하세요
titanic['who'].value_counts()# .plot(kind='bar')를 붙이면 위 sns로 했을때와 판다스로 썼을때와 똑같다.
```

## heatmap

```python
# 상품군 리스트
categories = ['의류', '가전', '식품', '생활용품', '뷰티', '스포츠', '도서']

# 월 리스트
months = ['1월', '2월', '3월', '4월', '5월', '6월', 
          '7월', '8월', '9월', '10월', '11월', '12월']

# 랜덤한 매출 데이터 생성
np.random.seed(42)
data = np.random.randint(50, 500, size=(len(categories), len(months)))

# DataFrame 생성
df_sales = pd.DataFrame(data, index=categories, columns=months)
df_sales
```

```python
# 👉 여기에 코드를 작성하세요
plt.figure(figsize=(12,7))
sns.heatmap(df_sales, cmap = 'Blues', annot=True, fmt='d')    #cmap(색상지정), annot(숫자로 표시해줘라), fmt=d(정수로 해줘)
```
